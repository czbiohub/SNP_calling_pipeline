```python
# read in this whole damn thing
vcf = VCF.dataframe('D12_B003528.vcf')
vcf
```

```python
gvcf = VCF.dataframe('D12_B003528.g.vcf')
gvcf
```

```python
# return a list corresponding to gvcf records that actually have coverage

def splitFunc(record):
    record_info = record['INFO']
    splt = record_info.split('END')
    myStr = splt[0]
    if myStr != "":
        return 1
    else: 
        return 0

keepList = gvcf.apply(splitFunc, axis=1)
keepList
```

```python
# filter gvcf by keepList
gvcf_filter = gvcf[np.array(keepList, dtype=bool)]
gvcf_filter
```

```python
# need to find the records that overlap between df and gvcf_filter
#     or rather, overlap in their CHROM and POS fields

merged = pd.merge(vcf, gvcf_filter, on='POS', suffixes=('_cellVCF', '_genomicVCF'))
merged
```

```python
merged.iloc[10]
```

```python
# lets give this a whirl
#    want stuff thats in gvcf ONLY
merge_diff = pd.merge(gvcf_filter, vcf, on='POS', how='outer', indicator=True, suffixes=('_cellVCF', '_genomicVCF'))
merge_diff
```

```python
# these are the records that have read coverage, and are WILDTYPE, ie. no variants
#     for a given cell
gvcf_ONLY_df = merge_diff[merge_diff['_merge'] == 'left_only']
gvcf_ONLY_df
```
