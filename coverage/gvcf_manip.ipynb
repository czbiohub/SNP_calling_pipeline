```python
# CRAZY FUCKIN FAST

def splitFunc(record):
    record_info = record['INFO']
    splt = record_info.split('END')
    myStr = splt[0]
    if myStr != "":
        return 1
    else: 
        return 0

gvcf = VCF.dataframe('D12_B003528.g.vcf')
keepList = gvcf.apply(splitFunc, axis=1)
keepList
```

```python
# define a list of just EGFR records -- think this works

def getEGFR(record):
    if record['CHROM'] == 'chr7':
        if 55207337 >= record['POS'] >= 55152337:
            return 1
        else:
            return 0
    else:
        return 0

toKeepList = df.apply(getEGFR, axis=1)
```

```python
# subset df by just our EGFR records
#      NO EGFR VARIANTS FOUND!
df_EGFR = df[np.array(toKeepList, dtype=bool)]
df_EGFR
```

```python
# do the same thing, for gvcf
#     NO EGFR RECORDS FOUND -- SUGGESTS NO COVERAGE TO EGFR!!
toKeepList_g = gvcf.apply(getEGFR, axis=1)
gvcf_EGFR = gvcf[np.array(toKeepList_g, dtype=bool)]
gvcf_EGFR
```

```python
# TEST CASE -- this should work. I know they both have this record

def getTEST_record(record):
    if record['CHROM'] == 'chr1':
        if 634115 >= record['POS'] >= 634110:
            return 1
        else:
            return 0
    else:
        return 0

toKeepList = df.apply(getTEST_record, axis=1)
toKeepList_g = gvcf.apply(getTEST_record, axis=1)

df_TEST = df[np.array(toKeepList, dtype=bool)]
gvcf_TEST = gvcf[np.array(toKeepList_g, dtype=bool)]
```

```python
# can we make this more generic? 

def getGOI_record(record, *args):
    chrom = 'chr' + str(args[0])
    start = args[1]
    end = args[2]
    
    if record['CHROM'] == chrom:
        if end >= record['POS'] >= start:
            return 1
        else:
            return 0
    else:
        return 0

chrom_ = 12
start_ = 6533927
end_ = 6538371

toKeepList = df.apply(getGOI_record, axis=1, args=(chrom_, start_ ,end_))
toKeepList_g = gvcf.apply(getGOI_record, axis=1, args=(chrom_, start_, end_))

df_TEST = df[np.array(toKeepList, dtype=bool)]
gvcf_TEST = gvcf[np.array(toKeepList_g, dtype=bool)]
```
