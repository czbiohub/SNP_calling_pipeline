```python
import pandas as pd
```

```python
cov_df = pd.read_csv('coverage_dataframe_all.csv')
cov_df
```

```python
# want to look at just our tumor cells here
tumor_cells_df = pd.read_csv('/Users/lincoln.harris/Desktop/LAUD_important_shit/tumor_cells_190429.csv', 
                             names=['index', 'cell'], header=0)
tumor_cells_df
```

```python
tumor_cells_list = list(tumor_cells_df['cell'])
tumor_cells_list
```

```python
# subset coverage_df by tumor cells
cov_df = cov_df[cov_df['cells'].isin(tumor_cells_list)]
cov_df
```

```python
cov_df = cov_df.reset_index(drop=True)
cov_df
```

```python
def split_func_v(x):
    """ returns variant count """
    if ':' in x: # not cellname
        outer = int(x.split(':')[0])
        inner = int(x.split(':')[1])
    
        ret = outer
    else:
        ret = x
        
    return(ret)



def split_func_t(x):
    """ returns total count """
    if ':' in x: # not cellname
        outer = int(x.split(':')[0])
        inner = int(x.split(':')[1])
    
        ret = inner
    else:
        ret = x
        
    return(ret)
```

```python
variant_cov_df = cov_df.applymap(split_func_v)
total_cov_df = cov_df.applymap(split_func_t)
variant_cov_df
```

```python
total_cov_df
```

```python
variant_cov_df.to_csv('variant_cov_df.csv', index=False)
total_cov_df.to_csv('total_cov_df.csv', index=False)
```

```python
meta = pd.read_csv('../metadata_all_cells_4.10.19.csv')
meta
```

```python
samples_set = set(meta['sample_name'])
samples_set
```

```python
# init dict to hold by-sample counts
samples_dict = dict.fromkeys(samples_set, 0)
samples_dict
```

```python
samples_dict['NOT_FOUND'] = 0
samples_dict
```

```python
# think this works, but pretty slow
for i in range(0,len(variant_cov_df.index)):
    currCell = variant_cov_df['cells'].iloc[i]
    keep = meta['cell_id'] == currCell
    meta_row = meta[keep]
    
    try:
        currSample = list(meta_row['sample_name'])[0]
    except IndexError: # cells not in metadata? 
        currSample = 'NOT_FOUND'
        
    for j in range(1,len(variant_cov_df.columns)): # starting at 1 bc i dont want the cellnames
        currVal = variant_cov_df.iloc[i,j]
        samples_dict[currSample] += currVal
```

```python
samples_pd = pd.DataFrame.from_dict(samples_dict, orient="index")
samples_pd
```
