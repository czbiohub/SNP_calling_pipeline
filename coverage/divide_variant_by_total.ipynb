```python
import pandas as pd
```

```python
# read in variant reads df, and sort gene names alphabetically
v = pd.read_csv('samples_x_gene_variant.csv')
v = v.set_index(v['gene'])
v = v.sort_index(ascending=True)
v
```

```python
# read in total reads df, and sort gene names alphabetically
t = pd.read_csv('samples_x_gene_total.csv')
t = t.set_index(t['gene'])
t = t.sort_index(ascending=True)
t
```

```python
# do they have the same genes?
v_genes = set(v.gene)
t_genes = set(t.gene)

print(len(v_genes))
print(len(t_genes))
```

```python
# appear to be exactly the same
v_genes.difference(t_genes)
```

```python
# sort colnames alphabetically?
v = v.reindex(sorted(v.columns), axis=1)
v
```

```python
# sort colnames alphabetically?
t = t.reindex(sorted(t.columns), axis=1)
t
```

```python
# they have the same colnames, right?
t_cols = set(t.columns)
v_cols = set(v.columns)

print(len(t_cols))
print(len(v_cols))

t_cols.difference(v_cols)
```

```python
# want to get rid of the 'gene' col, at the very end of each df
v = v.drop(['gene'], axis=1)
t = t.drop(['gene'], axis=1)
v
```

```python
t
```

```python
# this is so cool
div_df = v / t
div_df
```

```python
div_df = div_df.fillna(0)
div_df
```

```python
div_df.to_csv('gene_x_sample_ratios.csv', index=True)
```
