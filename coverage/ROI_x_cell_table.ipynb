```python
import os
import pandas as pd
#pd.set_option('display.max_rows', 999)
#pd.options.display.max_columns = None
```

```python
t = pd.read_csv('ROI_coverage/EGFR_L858R_coverageByCell.csv')
```

```python
cellNames = t['cellName']
cellNames
```

```python
files = os.listdir('ROI_coverage/')
files
```

```python
ROIs = []
for item in files:
    item_str = item.strip('_coverageByCell.csv')
    ROIs.append(item_str)

ROIs
```

```python
ROIs.insert(0, 'cellName')
ROIs
```

```python
big_df = pd.DataFrame(columns = ROIs)
big_df
```

```python
big_df['cellName'] = t['cellName']
big_df
```

```python
big_df['cellName']
```

```python
for file in files:
    ROI = file.strip('_coverageByCell.csv')
    #print(ROI)
    path = 'ROI_coverage/' + file
    curr = pd.read_csv(path)
    
    try:
        norm = curr['depth_vcf'] / curr['depth_gvcf']
        norm = norm.fillna(0)
    except TypeError: # ONLY ONE WIERD ONE, THANK GOD!!
        curr['depth_vcf'] = pd.to_numeric(curr['depth_vcf'], errors='coerce') 
        curr['depth_gvcf'] = pd.to_numeric(curr['depth_gvcf'], errors='coerce')

        norm = curr['depth_vcf'] / curr['depth_gvcf']
        norm = norm.fillna(0)
    
    # why are some of these values > 1.0? thats really fucking troubling
    for i in range(0, len(norm.index)):
        curr = norm.iloc[i]
        if curr > 1.0:
            norm.iloc[i] = 1.0
    
    big_df[ROI] = norm
```

```python
big_df.to_csv('ROI_x_cell.csv', index=False)
```
