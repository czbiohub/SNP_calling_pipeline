```python
import pandas as pd
import os
```

```python
goi_search_path = '/Users/lincoln.harris/code/SNP_calling_pipeline/getMutationCounts/out'
goi_files = os.listdir(goi_search_path)
goi_files
```

```python
for file in goi_files:
    gene = file.strip('_AA.csv')
    gene = gene.strip('_out')
    #print(gene)
    curr = pd.read_csv(goi_search_path + '/' + file, names=['cell', 'muts'])
    
    for i in range(0, len(curr.index)):
        currCell = curr['cell'].iloc[i]
        currMut = curr['muts'].iloc[i]
        
        if currMut != '[]':
            currMut = currMut.strip('[')
            currMut = currMut.strip(']')
            currMut = currMut.strip("'")
            
            # take care of the cells that have multiple hits
            currMutSplit = currMut.split(',')
            
            # how to iter list of unknown length? cool!   
            for i, elem in enumerate(currMutSplit):
                elem = currMutSplit[i].strip(" ")
                elem = elem.strip("'")
                #print(elem)
            
            ROI = gene + '_' + elem
            print(ROI)
```
