```python
# like a charm
def foo_bar(x):
    if ':' in x: # not cellname
        outer = int(x.split(':')[0])
        inner = int(x.split(':')[1])
            
        if inner != 0:
            ratio = outer / inner
        else:
            ratio = 0
        ret = ratio
        
    else:
        ret = x
    
    return(ret)
```

```python
cov_df = cov_df.applymap(foo_bar)
```
