```python
#/////////////////////////////////////////////////////////////////
#/////////////////////////////////////////////////////////////////
#
# what if i made a validationTable_samples, as opposed to 
#    validationTable_patients? 
#
#/////////////////////////////////////////////////////////////////
#/////////////////////////////////////////////////////////////////
```

```python
cols = ['sample', 'patient', 'driver_gene', 'driver_mutation', 'mutations_found', 'numCells', 'numTumorCells', 'numTumorCells_w_coverage_to_ROI', 'numTumorCells_clinMut_found']
validationTable_samples = pd.DataFrame(columns=cols)
validationTable_samples
```

```python
validationTable_samples['sample'] = relevantSamplesSeries
validationTable_samples
```

```python
# add patient
for i in range(0, len(validationTable_samples.index)):
    currSample = validationTable_samples['sample'][i]
    try:
        rowToKeep = patientMetadata['sample_name'] == currSample
        patientRows = patientMetadata[rowToKeep] # will return MULTIPLE rows
        patientRows = patientRows.reset_index(drop=True)

        patientID = patientRows['patient_id'][0]
        driverGene = patientRows['driver_gene'][0]
        driverMutation = patientRows['driver_mutation'][0]
        
        validationTable_samples['patient'][i] = patientID
        validationTable_samples['driver_gene'][i] = driverGene
        validationTable_samples['driver_mutation'][i] = driverMutation
    except:
        print('ERROR')
    
```
