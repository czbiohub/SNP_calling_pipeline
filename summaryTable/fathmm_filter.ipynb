```python
def get_laud_db():
    """ returns the COSMIC database after lung and fathmm filter """
    pSiteList = database.index[database['Primary site'] == 'lung'].tolist()
    database_filter = database.iloc[pSiteList]
    keepRows = database_filter['FATHMM score'] >= 0.7
    db_fathmm_filter = database_filter[keepRows]
    db_fathmm_filter = db_fathmm_filter.reset_index(drop=True)

    return db_fathmm_filter
```

```python
global database
global database_laud

database = pd.read_csv('../getMutationCounts/CosmicGenomeScreensMutantExport.tsv', sep='\t')
database_laud = get_laud_db()
database_laud
```

```python
mutsList = []

for i in range(0, len(database_laud.index)):
    row = database_laud.iloc[i]
    currGene = row['Gene name']
    currMut = row['Mutation AA']
    currMut = currMut.split('p.')[1]
    full = currGene + ' ' + currMut
    #print(full)
    mutsList.append(full)
```

```python
mutsList
```

```python
def mutationsDF_fillIn(GOI, GOI_df, mutationsDF_):
    mutName = GOI + '_mut'
    for i in range(0,len(mutationsDF_.index)):
        currCell = mutationsDF_['cell'][i]

        rightIndex = GOI_df['cell'] == currCell
        rightRow = GOI_df[rightIndex]

        rightCell = rightRow['cell']
        rightCell = str(rightCell).split()[1]
    
        rightMut = rightRow['mutations']
        rightMut = str(rightMut).split()[1]
        
        currMut = ''.join(rightMut)
        currMut = currMut.replace("'", "")
        currMut = currMut.replace("]", "")
        currMut = currMut.replace("[", "")
        currMut = currMut.replace(" ", "")    

        mutStr = GOI + ' ' + currMut

        if mutStr in mutsList:
            print('got a match')
            mutationsDF_[mutName][i] = currMut
        else:
            mutationsDF_[mutName][i] = ''
```

```python
for gene in genesList:
    gene_path = muts_path + gene + '_out_AA.csv'
    gene_df = pd.read_csv(gene_path, header=None, names=['cell', 'mutations'])
    mutationsDF_fillIn(gene, gene_df, mutationsDF)
    #summarizeModule.removeExtraCharacters_mutationsDF(gene, mutationsDF)
```
