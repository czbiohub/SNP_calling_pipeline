```python
# lets turn this into a function 
def addCoverageFunc(coverage_df, queryGene, queryMutation):
    for i in range(0, len(validationTable_cells.index)):
        currCell = validationTable_cells['cell'][i]
        currDriver = validationTable_cells['clinical_driver_gene'][i]
        currMut = validationTable_cells['clinical_mutation'][i]
    
        if currDriver == queryGene and currMut == queryMutation:
            if currCell in list(coverage_df['cellName']):
                index_cov_nonZero = coverage_df['cellName'] == currCell
                currRow_cov_nonZero = coverage_df[index_cov_nonZero]
                currDepth_gvcf = int(currRow_cov_nonZero['depth_gvcf'])
        
                validationTable_cells['coverage_to_ROI'][i] = currDepth_gvcf
            else:
                validationTable_cells['coverage_to_ROI'][i] = 0
```
