```python
def get_bam(cell):
    s3_location = f'{prefix}{cell}'
    lines = ! aws s3 ls $s3_location
    bam_line = [x for x in lines if x.endswith('bam')][0]
    bam_basename = bam_line.split()[-1]
    return f'{s3_location}{bam_basename}'


cells['input_bam'] = cells['cell_name'].map(get_bam)
cells.head()
```

```python
cells['sample_id'] = cells.cell_name.str.strip('/')
cells['id'] = cells['sample_id']
cells.head()
```

```python
import os

cells['output_prefix'] = cells['input_bam'].map(os.path.dirname)
cells.head()
```

```python
cells['output_vcf'] = cells.apply(
    lambda x: '{output_prefix}/{sample_id}.vcf'.format(**x), axis=1)
cells.head()
```

```python
columns = ['id', 'input_bam', 'sample_id', 'output_vcf']

samples = cells[columns]
samples
```
