```python
for item in scVCF_list:
    currCell = item.strip('.vcf')
    currPlate = currCell.split('_')[1]
    
    rowToKeep = patientMetadata['plate'] == currPlate
    
    try:
        currPatient = patientMetadata['patient_id'][rowToKeep]
        currPatientVal = currPatient.item()
        currDictVal = patientsDict[currPatientVal]
        toUpdateVal = currDictVal + 1
        patientsDict.update({currPatientVal:toUpdateVal})
    except:
        print('ERROR')
```

```python
# what patients do we have a bunch of cells from? 
#     TH226 looks like a good choice
patientsDict
```

```python
# lets get a list of the cells that come from TH179
TH179_cells = []

for item in scVCF_list:
    currCell = item.strip('.vcf')
    currPlate = currCell.split('_')[1]
    
    rowToKeep = patientMetadata['plate'] == currPlate
    
    try:
        currPatient = patientMetadata['patient_id'][rowToKeep]
        currPatientVal = currPatient.item()
        if currPatientVal == 'TH179':
            TH179_cells.append(currCell)
    except:
        continue
```

```python
TH179_files = []
fullPath = '/Users/lincoln.harris/code/SNP_calling_pipeline/bulkAnalysis/scVCF/'
for item in TH179_cells:
    file = fullPath + item + '.vcf'
    TH179_files.append(file)
```
