```python
# lets turn this into a function
import os
import pandas as pd
pd.options.display.max_colwidth = 500 # module config? 
pd.options.mode.chained_assignment = None  # disable warning message? 
    
# getSampleID()
#     TODO ADD DESCRIPTION
def getSampleID(cell):
    split = cell.split('_')
    currCell = split[0] + '_' + split[1]
    return currCell

# driver()
#   TODO ADD DESCRIPTION
def driver(prefix):
    # get the bam and bam index files
    f = 'myFile.txt'
    ! aws s3 ls $prefix > $f
    
    # read run names into a dataframe
    files_df = pd.read_table(f, delim_whitespace=True, header=None, names=['date','time','size','fileName'])
    
    # get a full_path col
    files_df['full_path'] = prefix + files_df['fileName']

    # get rid of .bai files -- more efficient (hopefull this is ok to do...)
    files_df1 = files_df.iloc[::6]
    
    # set sample_id to dummy variable
    files_df1['sample_id'] = files_df1['time']
    
    # add sampleID col
    files_df1['sample_id'] = files_df1['fileName'].map(getSampleID)

    # add id col
    files_df1['id'] = files_df1['sample_id']
    
    # keep only the cols we care about
    cols_to_keep = ['id', 'full_path', 'sample_id']
    files_df2 = files_df1[cols_to_keep]
    files_df2['input_bam'] = files_df2['full_path']

    # building the output vcf string
    files_df2['output_vcf'] = files_df2['input_bam'].map(os.path.dirname)
    
    # add seperate file extension for vcfs
    files_df2['output_vcf'] = files_df2['output_prefix'] + '/vcf/'
    
    # get cell names into vcfs? -- perfect
    files_df2['output_vcf'] = files_df2['output_vcf'] + files_df2['id'] + '.vcf'
   
    # subset again
    cols_to_keep = ['id', 'input_bam', 'sample_id', 'output_vcf']
    files_df3 = files_df2[cols_to_keep]
    
    return files_df3
```
