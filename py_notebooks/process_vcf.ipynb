```python
import os

files = []
for file in os.listdir("/Users/lincoln.harris/Desktop/vcf/all"):
    if file.endswith(".vcf"):
        fullPath = (os.path.join("/Users/lincoln.harris/Desktop/vcf/all/", file))
        files.append(fullPath)
```

```python
files
```

```python
cells_dict = {}

for f in files:
    cell = f.replace("/Users/lincoln.harris/Desktop/vcf/all/", "")
    cell = cell.replace(".vcf", ""
                       )
    vcf_rdr = vcf.Reader(open(f, 'r'))
    unique = []
    
    for record in vcf_rdr:
        currPos = record.POS
        if currPos not in unique:
            unique.append(currPos)
    
    cells_dict.update({cell, len(unique)})
    
```

```python
import csv

with open('myfile.csv', 'wb') as outfile:
    w = csv.writer(outfile)
    w.writerows(cells_dict.items())
```
