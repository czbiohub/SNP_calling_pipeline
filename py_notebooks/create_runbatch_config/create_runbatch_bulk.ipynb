```python
import pandas as pd
```

```python
# get all of the cells in a given run directory
prefix = 's3://darmanis-group/singlecell_lungadeno/rawdata/bulk_bams/'
txt = 'bulk.txt'
! aws s3 ls $prefix > $txt
! cat $txt
```

```python
# read 180226 cell names into a dataframe
s3_df = pd.read_table(txt, delim_whitespace=True, header=None, names=['date', 'time', 'size','file'])
s3_df
```

```python
s3_df = s3_df.drop([0])
s3_df
```

```python
patients_df = pd.DataFrame(data=s3_df['file'], columns=['file'])
patients_df
```

```python
patients_df['patient_ID'] = ''
patients_df
```

```python
for i in range(0,len(patients_df.index)):
    currFile = patients_df['file'].iloc[i]
    currFileSplit = currFile.split('_')
    currID = currFileSplit[0] + currFileSplit[1]
    currID_split = currID.split('E')[0]
    patients_df['patient_ID'].iloc[i] = currID_split

patients_df
```

```python
patients_df['sample_ID'] = ''
patients_df
```

```python
# how many of these CBC guys are there? 
for i in range(0,len(patients_df.index)):
    currFile = patients_df['file'].iloc[i]
    currFileSplit = currFile.split('_')
    
    if 'CBC' in currFile: 
        print(currFile)
```

```python
for i in range(0,len(patients_df.index)):
    currFile = patients_df['file'].iloc[i]
    currFileSplit = currFile.split('_')
    
    if currFileSplit[1].isnumeric():
        currID = currFileSplit[0] + currFileSplit[1] + '_' + currFileSplit[2] + '_' + currFileSplit[3]
    elif 'L001' in currFileSplit[2]:
        currID = currFileSplit[0] + '_' + currFileSplit[1]
    else:
        currID = currFileSplit[0] + '_' + currFileSplit[1] + '_' + currFileSplit[2]
        
    #print(currID)
    patients_df['sample_ID'].iloc[i] = currID

patients_df
```

```python
patients_df['s3_path'] = prefix + patients_df['file']
patients_df
```

```python
patients_df['s3_path'][10]
```

```python
outPath = 's3://darmanis-group/singlecell_lungadeno/non_immune/nonImmune_bams_9.27/bulk_vcf1/'
```

```python
suffix = 'homo.Aligned.out.sorted.bam'
colNames = ['id', 'input_bam', 'sample_id', 'output_vcf']
toAddRows_list = []

for i in range(1,len(patients_df.index)):
    fName = patients_df['file'].iloc[i]
    if fName.endswith(suffix):
        id_ = patients_df['sample_ID'].iloc[i]
        input_bam_ = patients_df['s3_path'].iloc[i]
        sample_id_ = patients_df['sample_ID'].iloc[i]
        output_vcf_ = outPath + id_ + '.vcf'
        print(output_vcf_)
        toAddRow = pd.DataFrame([[id_, input_bam_, sample_id_, output_vcf_]], columns=colNames)
        toAddRows_list.append(toAddRow)
        
```

```python
toAddRows_list
```

```python
outputDF = pd.concat(toAddRows_list, ignore_index=True)
outputDF
```
