```python
import os

files = []
for file in os.listdir("/home/ubuntu/expansionVol2/04-GATK_out/all"):
    if file.endswith(".vcf"):
        fullPath = (os.path.join("/home/ubuntu/expansionVol2/04-GATK_out/all/", file))
        files.append(fullPath)
```

```python
files
```

```python
cells_dict = {}

for f in files:
    cell = f.replace("/home/ubuntu/expansionVol2/04-GATK_out/all/", "")
    cell = cell.replace(".vcf", ""
                       )
    vcf_rdr = vcf.Reader(open(f, 'r'))
    unique = []
    
    for record in vcf_rdr:
        currPos = record.POS
        if currPos not in unique:
            unique.append(currPos)
    
    #print(len(unique))
    val = len(unique)
    cells_dict.update({cell : val})
    
```

```python
cells_dict
```

```python
import csv

with open('myfile.csv', 'wb') as outfile:
    w = csv.writer(outfile)
    w.writerows(cells_dict.items())
```
