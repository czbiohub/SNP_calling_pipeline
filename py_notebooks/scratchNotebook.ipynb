```python
import pandas as pd
foo = pd.read_csv('./foo.csv', index_col=0)
#foo = pd.DataFrame(foo)
foo
```

```python
cellNames = list(foo.index)
cellNames
```

```python
genesList = []
for i in range(0,foo.shape[0]):
    currList = list(foo.iloc[i].unique()) # unique genes for curr_cell 
    
    for elm in currList:
        if elm not in genesList:
            genesList.append(elm)

genesList1 = pd.Series(genesList)
genesList1 # all genes in dataset!!
```

```python
df = pd.DataFrame(columns=genesList1, index=cellNames)
df
```

```python
# set everybody to zero
for col in df.columns:
    df[col] = 0
df
```
